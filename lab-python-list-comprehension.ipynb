```python
def initialize_inventory(products):
    inventory = {producto : int(input(f"Enter the quantity of {producto} available:" ))for producto in products}
    return inventory
```

```python

def get_customer_orders(lista_productos):
    number_customer_products = int(input("Enter the number of customer orders: "))

    raw_orders = [
        input("Enter the name of a product that a customer wants to order: ")
        for _ in range(number_customer_products)
    ]

    valid_orders = [p for p in raw_orders if p in lista_productos]

    return valid_orders

```

```python


def update_inventory(customer_orders, inventory):

    updated = {producto: (inventory[producto] - 1 if producto in customer_orders else inventory[producto]) for producto in inventory}

    updated = { producto: cantidad for producto, cantidad in updated.items() if cantidad > 0}

    [print(f"{producto} : {cantidad}") for producto, cantidad in updated.items()]

    return updated
```

```python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_unique = round((total_products_ordered / len(products)) * 100, 2)
    return total_products_ordered, percentage_unique

```

```python
def print_order_statistics(order_statistics):
    valor1, valor2 = order_statistics
    print(f"Total Products Ordered: {valor1}")
    print(f"Percentage of Unique Products Ordered: {valor2}")

```

```python
def calculate_total_price(customer_orders):
    
    inventario_precios = {producto: float(input(f"Enter the price of {producto}: ")) for producto in customer_orders}
    factura_cliente = sum(inventario_precios.values())

    
    return factura_cliente
```

```python
lista_productos = ['t-shirt','mug', 'hat', 'book', 'keychain']
```

```python

inventory = initialize_inventory(lista_productos)
```

```python
customer_orders = get_customer_orders(lista_productos)
```

```python
order_statistics = calculate_order_statistics(customer_orders, lista_productos)
```

```python
print_order_statistics(order_statistics)
```

```python
print(f"Updated Inventory:")
inventory = update_inventory(customer_orders, inventory)
```

```python
print(f"Total Price: {calculate_total_price(customer_orders)}")
```
